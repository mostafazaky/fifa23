```python
import numpy as np
import pandas as pd 

# for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
working_dirctory= os.getcwd()

print(working_dirctory)
```

```python
path = working_dirctory + '/data.csv'
fifa = pd.read_csv(path)
```

```python
fifa.head()
```

```python
fifa.info()
```

```python
fifa.describe()
```

```python
def country(C):
    return fifa[fifa['Nationality']== C][['Name', 'Overall', 'Potential', 'Position']]

country('Egypt')
```

```python
def club(x):
    return fifa[fifa['Club'] == x][['Name','Position','Overall','Nationality','Kit Number','Age','Wage',
                                    'Value','Preferred Foot']]

club('Liverpool')
```

```python
plt.figure(figsize = (7,7))
pl = sns.boxplot(x=fifa['Club'][1:50], y=fifa['Overall'], palette='hls');
pl.set_title(label='Overall - Club');
plt.xlabel('Club')
plt.ylabel('Overall')
plt.xticks(rotation = 80)
plt.ylim(85, 95)
```

```python
players = fifa[['Name','Work Rate',
                'Skill Moves','Preferred Foot','Wage','Weak Foot']]

players.head(7)
```

```python
fifa[['Skill Moves', 'Name','Club']].sort_values('Skill Moves', ascending=False).nlargest(20, 'Skill Moves')
```

```python
plt.rcParams['figure.figsize'] = (19, 5)
sns.countplot(fifa['Skill Moves'], palette = 'pink')
plt.title('Skills', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (13, 9))
ax = sns.countplot(x = 'Height', data = fifa, palette = 'dark')
ax.set_title(label = 'Count of players on Basis of Height', fontsize = 20)
ax.set_xlabel(xlabel = 'Height in Foot per inch', fontsize = 10)
ax.set_ylabel(ylabel = 'Count', fontsize = 16)
plt.show()
```

```python
#As of curiosity, I want to know the tallest five players
fifa[['Height', 'Name','Club']].sort_values('Height', ascending=True).head(5)
```

```python
#As of curiosity, I want to know the tallest five players
fifa[['Height', 'Name','Club']].sort_values('Height', ascending=False).head(5)
```

```python
plt.style.use('dark_background')
fifa['Nationality'].value_counts().head(40).plot.bar(color = 'red', figsize = (20, 7))
plt.title('Different Nations Participating in FIFA 2023', fontsize = 35, fontweight = 25)
plt.xlabel('Name of The Country')
plt.ylabel('count')
plt.show()

```

```python
sns.set_theme()
df = sns.load_dataset("penguins")
sns.set(style = "dark", palette = "pink", color_codes = True)
x = fifa.Age
plt.figure(figsize = (15,8))
ax = sns.distplot(x, bins = 58, kde = False,hist_kws={'linewidth': 3, 'linestyle': '-'})
ax.set_xlabel(xlabel = "Player\'s age", fontsize = 16)
ax.set_ylabel(ylabel = 'Number of players', fontsize = 16)
ax.set_title(label = 'Histogram of players age', fontsize = 20)
plt.show()
```
